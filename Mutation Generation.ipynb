```python
import pandas as pd
import glob
import numpy as np

path = r'./data/' # use your path
all_files = glob.glob(path + "/*.txt")

li = []

for filename in all_files:
    if filename[7:-9]== 'khaghani':
        continue
    df = pd.read_csv(filename, names=['mesra'], header=None)
    df['beyt index'] = df.index // 2
    df = df.groupby(['beyt index']).agg({'mesra': ' '.join})
    df = df.rename(columns={'mesra':'beyt'})
    df['author'] = filename[7:-9]
    li.append(df)

df = pd.concat(li, axis=0, ignore_index=True)
```

```python
import gensim

# NLP stuff
import string
from hazm import *
import codecs
```

```python
normalizer = Normalizer()
stopwords = [normalizer.normalize(x.strip()) for x in codecs.open('stopwords.dat','r','utf-8').readlines()]
lemmatizer = Lemmatizer()

def preprocess(text):
    tokens = word_tokenize(text)
    tokens = [lemmatizer.lemmatize(t) for t in tokens if t not in stopwords]
    return tokens
```

```python
data_preprocessed = df['beyt'].apply(preprocess)
data_preprocessed = data_preprocessed.tolist()
```

```python
import logging
logging.basicConfig(filename='gensim.log',
                    format="%(asctime)s:%(levelname)s:%(message)s",
                    level=logging.INFO)
```

```python
from gensim.models import FastText

model_ted = FastText(data_preprocessed, size=100, window=5, min_count=5, workers=11,sg=1)
```
